```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('class data.csv').dropna()
df.head()
```

```python
df.plot.box()
```

```python
pd.plotting.scatter_matrix(df,figsize=(16,6))
```

```python
form='Aim_pct ~ I(Water)+Hrs_sleep+I(Siblings)+I(Selfies)'
model = sm.formula.ols(data = df,formula=form).fit()
```

```python
model.summary()
```

```python
model.resid.plot.hist()
```

```python
sns.residplot(model.fittedvalues, model.resid)
```
