```python
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('airline_passengers.csv').drop(144,axis='index')
df.head()
```

```python
df.index = pd.to_datetime(df['Month'], format='%Y-%m')
df = df.drop('Month', axis='columns')
df.head()
```

```python
df.plot.line(title='Airline passenges plot', figsize=(16,5))
```

# hodrick prescott filter

```python
cycle, trend = sm.tsa.filters.hpfilter(df)
df['cycle'] = cycle
df['trend']= trend
df.plot(figsize=(16,6), title='hodrick presoct fillter')
```

# seasonal decompsition

```python
fig =sm.tsa.seasonal_decompose(df['Thousands of Passengers']).plot()
```
