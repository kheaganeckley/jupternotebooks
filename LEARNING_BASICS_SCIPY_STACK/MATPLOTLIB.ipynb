```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data ={
    'a' : [1,2,3,4,5,5,6],
    'b' : [2,3,4,5,66,7,7]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.to_csv('hello') # make a csv from data frame data
```

```python
df.plot(kind='density', title='kheagans amazing plot', linestyle='-', linewidth=3,)
```

```python
df['a'].plot(kind='line', title='a line graph')
```

```python
df.plot(kind='box', title='kheagans box plot', ylim=(0,7.5))
```

```python
df.plot(kind='area')
```

```python
df.plot.scatter(x='a', y='b')
```

```python
df.plot.bar()
```

```python
df.plot.pie(subplots=True)
```
