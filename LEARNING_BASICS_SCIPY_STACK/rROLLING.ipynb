# rolling mean
note how the two graphs differ

```python
import pandas as pd
```

```python
df = pd.read_csv('walmart_stock.csv', index_col='Date', parse_dates=True)
df.head()
```

```python
df['Open'].plot.line(figsize=(16,6))
```

```python
dfRolling =df.rolling(7).mean()
dfRolling.head(10)
```

```python

df['Moving avrage 30 days'] = df['Close'].rolling(30).mean()
df[['Moving avrage 30 days', 'Close']].plot.line(figsize=(16,6))
```
